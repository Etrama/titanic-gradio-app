```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings
warnings.filterwarnings('ignore')
import config
```

```python
train = pd.read_csv(config.raw_train_data)
train = train.drop(['PassengerId', "Name", "Ticket", "Cabin"], axis=1)
train = pd.get_dummies(data=train, columns=["Embarked", "Sex"], drop_first=True)
# leaving Pclass, SibSp and Parch for later
convert_columns = ["Embarked_Q", "Embarked_S", "Sex_male",]
for column in convert_columns:
    train[column] = train[column].astype("int64")
```

```python
train["Age"] = train["Age"].fillna(train["Age"].mean())
```

```python
train.to_csv(config.processed_train_data, index=False)
```
